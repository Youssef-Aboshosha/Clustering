```python
# Importing the required libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
# Generating sample data
np.random.seed(0)
X = np.concatenate((np.random.randn(100, 2) * 0.5 + [2, 2],
                    np.random.randn(100, 2) * 0.5 + [-2, -2],
                    np.random.randn(100, 2) * 0.5 + [2, -2]))
```

```python
# Visualizing the sample data
plt.scatter(X[:, 0], X[:, 1], c='b', alpha=0.5)
plt.title('Sample Data')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
# Creating a K-means clustering model with 3 clusters
kmeans = KMeans(n_clusters=3, random_state=0)
```

```python
# Fitting the model to the data
kmeans.fit(X)
```

```python
# Getting the cluster labels and cluster centers
labels = kmeans.labels_
centers = kmeans.cluster_centers_
```

```python
# Visualizing the clustered data
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', alpha=0.5)
plt.scatter(centers[:, 0], centers[:, 1], c='r', marker='x', s=200)
plt.title('Clustered Data')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```
